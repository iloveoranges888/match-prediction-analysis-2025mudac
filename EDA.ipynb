### Vizualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your dataframe
plt.figure(figsize=(8, 6))
sns.histplot(df['Match Length'], kde=True)
plt.title('Distribution of Match Length')
plt.xlabel('Match Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender distribution for Bigs
big_gender_counts = df['Big Gender'].value_counts()

# Gender distribution for Littles
little_gender_counts = df['Little Gender'].value_counts()

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
big_gender_counts.plot(kind='bar', ax=ax[0], title='Gender Distribution of Bigs')
little_gender_counts.plot(kind='bar', ax=ax[1], title='Gender Distribution of Littles')

ax[0].set_ylabel('Count')
ax[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Race/Ethnicity distribution for Bigs
big_race_counts = df['Big Race/Ethnicity'].value_counts()

# Race/Ethnicity distribution for Littles
little_race_counts = df['Little Participant: Race/Ethnicity'].value_counts()

# Plotting stacked bar chart
race_df = pd.DataFrame({
    'Bigs': big_race_counts,
    'Littles': little_race_counts
}).fillna(0)

race_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Race/Ethnicity Distribution for Bigs and Littles')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot of Program Type by Match Stage
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Program Type', hue='Stage')
plt.title('Program Type by Match Stage')
plt.xlabel('Program Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
closure_reason_counts = df['Closure Reason'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=closure_reason_counts.index, y=closure_reason_counts.values)
plt.title('Match Closure Reasons')
plt.xlabel('Closure Reason')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 closure reasons sorted by count
top_10_closure_reasons = df['Closure Reason'].value_counts().sort_values(ascending=False).head(10)

# Plotting a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_closure_reasons.index, y=top_10_closure_reasons.values)
plt.title('Top 10 Match Closure Reasons')
plt.xlabel('Closure Reason')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming 'Big Days Interview to Acceptance' contains the number of days
plt.figure(figsize=(8, 6))
sns.histplot(df['Big Days Interview to Acceptance'], kde=True)
plt.title('Distribution of Days Between Interview and Acceptance')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert the date column to datetime format
df['Big Enrollment: Created Date'] = pd.to_datetime(df['Big Enrollment: Created Date'])

# Group by month and count enrollments
enrollment_counts = df['Big Enrollment: Created Date'].dt.to_period('M').value_counts().sort_index()

# Plotting the time series
enrollment_counts.plot(kind='line', figsize=(10, 6))
plt.title('Enrollments by Date')
plt.xlabel('Month')
plt.ylabel('Number of Enrollments')
plt.show()
```

```python
# Categorize ages into groups
bins = [0, 18, 30, 40, 50, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51+']
df['Age Group'] = pd.cut(df['Big Age'], bins=bins, labels=labels, right=False)

# Plotting the bar chart for age group distribution
age_group_counts = df['Age Group'].value_counts()
age_group_counts.plot(kind='bar', figsize=(8, 6))
plt.title('Program Participation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 most frequent Big Employers
top_10_employers = df['Big Employer'].value_counts().head(10)

# Plotting a bar chart for the top 10 Big Employers
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_employers.index, y=top_10_employers.values)
plt.title('Top 10 Big Employers')
plt.xlabel('Employer')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count how many Bigs are marked as unemployed (assuming 'Unemployed' is in the "Big Employer" column)
unemployed_count = df[df['Big Employer'].str.contains('Unemployed', case=False, na=False)].shape[0]

# Print the result
print(f'Number of unemployed Bigs: {unemployed_count}')
```
